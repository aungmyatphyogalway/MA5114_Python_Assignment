```python
##(1)
from Bio import Entrez
from Bio import SeqIO

try:
    Entrez.email = "A.N.Other@example.com"   #Note: if you don't specify an email address for this session, a warning will be generated
    handle = Entrez.efetch(db="nucleotide", rettype="gb", retmode="text", id="NG_008720")
    hfe = SeqIO.read(handle, "gb")
    handle.close()
    print(type(hfe))
except:
    print('Failed to fetch record.')
```

```python
sequence=hfe.seq
print(sequence)
```

```python
##(2)
def get_cds_feature_with_qualifier_value(seq_record):
    """Function to look for CDS feature by annotation value in sequence record.
    
    e.g. You can use this for finding features by locus tag, gene ID, or protein ID.
    """
    # Loop over the features
    for feature in seq_record.features:
        if feature.type == "CDS":
            return feature
    # Could not find it
    return None


cds_feature = get_cds_feature_with_qualifier_value(hfe)
print(cds_feature)
```

```python
gene_sequence = cds_feature.extract(hfe.seq)
print("CDS nucleotide sequence:")
print(gene_sequence)
```

```python
##(3)
protein_sequence = gene_sequence.translate(table=11, cds=True)
print("Translated into amino acids:")
print(protein_sequence)
```

```python
import re
```

```python
hinfl=re.compile(r'GANTC')
rsal=re.compile(r'GTAC')
ecori=re.compile(r'GAATTC')
```

```python
def enzyme(enz,seq):
    if re.search(enz,seq):
        print('restriction_found')
    else:
        print('no restriction site found')
```

```python
enzyme(hinfl,gvariant)
enzyme(hinfl,avariant)
```

```python
enzyme(rsal,gvariant)
enzyme(rsal,avariant)
```

```python
enzyme(ecori,gvariant)
enzyme(ecori,avariant)
```

```python
enzyme(hinfl,gg)
enzyme(hinfl,aa)
```

```python
enzyme(rsal,gg)
enzyme(rsal,aa)
```

```python
enzyme(ecori,gg)
enzyme(ecori,aa)
```
