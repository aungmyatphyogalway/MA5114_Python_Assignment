```python
###Question2
seqq=' '
with open('genomic_RNA.fa','r')as rna:
    for line in rna:
        if line.startswith('>'):
            r=line.split()
        else:
            s=line.split()
            seqq=s
            print(seqq)
```

```python
seq=str(seqq)
with open('exons.txt','r')as ent:
    ex=ent.read()   
exons=ex.split() 
print(exons)
```

```python
##Q2(i)and(ii)
mrna=" "
for exon in exons:
    start=int(exon.split(',')[0])
    stop=int(exon.split(',')[1])
    exon_seq= seq[start:stop]
    mrna=mrna+exon_seq
print(mrna)    
```

The critical thing is that the DNA reads the same way from either side - whether its from 5'-3' or 3'-5', we read A-C-G-C-G-T. It is an interesting - and important problem - to study these sequences, and their properties, as they form the basis for a defence mechanism that has evolved in bacteria and that has been recently adopted as a cutting edge way of manipulating the genome, called CRISPR.

For this final part of the assignment, you will find the complete human reference mitochondrial chromosome, chrM.fa. You will need to think about the following:

```python
##Q3
with open('chrM.fa','r')as hu:
    a=hu.read()
    print(a)
    
```

```python
dna=a.upper()
```

```python
def complement(sequence):
    transcript_dict = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    complement = ''
    for base in sequence:
        complement += transcript_dict[base]
    return complement
```

```python
def reverse_sequence(sequence):
    reversed_seq = ''
    seq_as_list = list(sequence)
    for base in reversed(seq_as_list):
        reversed_seq += base
    return reversed_seq
```

```python
def reverse_complement(sequence):
    complement_seq = complement(sequence)
    reverse_complement = reverse_sequence(complement_seq)
    return reverse_complement
```
